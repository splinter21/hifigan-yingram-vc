```python
import torch
```

```python
from hubconf import hifigan_wavlm as hifigan
from hubconf import get_full_features, get_hubert_model

import soundfile
```

```python
dec=hifigan(model_path="g_00444000.pt",device="cpu")
hubert = get_hubert_model()
```

```python
from pathlib import Path

src_wav_path = Path('君の知らない物語-src.wav')
query_seq = get_full_features(src_wav_path,hubert,device="cpu", shift=-1)
# prediction = self.vocode(out_feats[None].cuda()).cpu().squeeze()
with torch.no_grad():
    y_g_hat = dec[0](query_seq.unsqueeze(0))
    y_g_hat = y_g_hat.squeeze(1).squeeze(0)
soundfile.write("out.wav", y_g_hat.cpu().numpy(), 32000)
import IPython.display as ipd
ipd.Audio("out.wav") # load a local WAV file
```
